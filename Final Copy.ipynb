### Data Cleaning

```python
# Select specific columns by name
selected_columns = df[['Reporting Railroad Code', 'Reporting Railroad Name', 'Year', 'Accident Year', 
                       'Accident Month', 'Date', 'Accident Type', 'Division', 'Station', 'State Name', 
                       'Equipment Type', 'Train Speed', 'Recorded Estimated Speed', 'Total Damage Cost', 
                       'Primary Accident Cause', 'Hours Engineers On Duty', 'Hours Conductors On Duty']]

# Create a new DataFrame containing only the selected columns
cleaned_df = selected_columns.copy() 

# Drop rows with Na values in any column
cleaned_df = selected_columns.dropna()

cleaned_df
```

```python
# Save the cleaned data to a CSV file for better exploration 
cleaned_df.to_csv('final_project_data.csv', index=False)
```

### 1. How has the frequency of train accidents changed over the years for New Jersey Transit Rail Operations?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the dataset for New Jersey Transit Rail Operations
selected_company = 'New Jersey Transit Rail Operations'
filtered_data = cleaned_df[cleaned_df['Reporting Railroad Name'] == selected_company]

# Group the filtered data by 'Year' and count the number of accidents each year
accidents_by_year = filtered_data.groupby('Year').size()

# Plotting the trend
plt.figure(figsize=(10, 6))
accidents_by_year.plot(kind='line', title=f'Train Accidents Trend for {selected_company} Over the Years', ylabel='Number of Accidents', xlabel='Year', marker='o', linestyle='-', color='b')

# Adding gridlines for better visualization
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

This code generates a line graph that shows the trend in the frequency of train accidents for New Jersey Transit Rail Operations over the years. 

### 2. Which stations have the highest number of train accidents? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the railroad of interest
selected_railroad = "New Jersey Transit Rail Operations"

# Filter the data for the selected railroad
selected_railroad_data = cleaned_df[cleaned_df['Reporting Railroad Name'] == selected_railroad]

# Group the data by 'Station Name' and count the number of accidents for each station
station_accident_counts = selected_railroad_data['Station'].value_counts()

# Select the top stations with the highest number of accidents (adjust the number as needed)
top_stations = station_accident_counts.head(10)

# Create a scatter plot to visualize the top stations with dots and labels
plt.figure(figsize=(12, 8))
plt.scatter(top_stations.index, top_stations.values, c='b', marker='o', s=100)  # Dots for stations
plt.title(f'Top Stations with Highest Number of Train Accidents ({selected_railroad})')
plt.xlabel('Station Name')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Attach station names as labels
for station, count in zip(top_stations.index, top_stations.values):
    plt.annotate(count, (station, count), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=10)

plt.show()
```

### 3. Is there a correlation between the Hours Engineers On Duty and accidents at New Jersey Transit Rail Operations?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selected company
selected_company = ['New Jersey Transit Rail Operations']

# Reapplying the filter to the original dataset for the selected company
selected_company_data = cleaned_df[cleaned_df['Reporting Railroad Name'].isin(selected_company)]

# Filtering for the top 10 causes for the selected company
top_causes_selected_company_reduced = selected_company_data['Primary Accident Cause'].value_counts().head(10).index.tolist()

# Create a filtered DataFrame for the selected company and top causes
filtered_data_selected_company_reduced = selected_company_data[selected_company_data['Primary Accident Cause'].isin(top_causes_selected_company_reduced)]

# Create a line plot with a smooth curve for 'Hours Engineers On Duty' by 'Primary Accident Cause'
plt.figure(figsize=(12, 8))
sns.lineplot(x='Primary Accident Cause', y='Hours Engineers On Duty', data=filtered_data_selected_company_reduced, err_style='bars')
plt.title('Smooth Line Plot of Hours Engineers On Duty by Primary Accident Cause')
plt.xlabel('Primary Accident Cause')
plt.ylabel('Hours Engineers On Duty')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

### 4. What are the most common causes of train accidents for New Jersey Transit Rail Operations?

```python
import matplotlib.pyplot as plt

# Selected companies
selected_company = ['New Jersey Transit Rail Operations']

# Filter the dataset for the selected companies
filtered_data_for_pie = cleaned_df[cleaned_df['Reporting Railroad Name'].isin(selected_company)]

# Grouping the filtered data by company and finding the top 10 causes for each
top_10_causes_by_company_for_pie = filtered_data_for_pie['Primary Accident Cause'].value_counts().head(10)

# Resetting index to convert it into a DataFrame
top_10_causes_by_company_df_for_pie = top_10_causes_by_company_for_pie.reset_index(name='Count')
top_10_causes_by_company_df_for_pie.rename(columns={'index': 'Cause'}, inplace=True)

# Setting up the figure for pie charts with an adjusted size
fig, axes = plt.subplots(1, len(selected_company), figsize=(18, 7))

# Ensure axes is a list if only one company is selected
if not isinstance(axes, list):
    axes = [axes]

# Creating a pie chart for each company
for i, (company, ax) in enumerate(zip(selected_company, axes)):
    # Filter data for the company
    company_data = top_10_causes_by_company_df_for_pie
    
    # Define a color palette
    palette = plt.cm.Paired(range(len(company_data)))

    # Create the pie chart
    wedges, texts, autotexts = ax.pie(
        company_data['Count'],
        labels=company_data['Cause'],
        autopct='%1.1f%%',
        startangle=140,
        colors=palette
    )
    
# Add a common legend for all pie charts
fig.legend(
    labels=top_10_causes_by_company_df_for_pie['Cause'],
    title="Causes",
    loc="upper right",
    bbox_to_anchor=(1.2, 1)
)

fig.suptitle('Proportion of Top 10 Accident Causes for New Jersey Transit Rail Operations', fontsize=16)
plt.subplots_adjust(wspace=0.5)  # Adjust horizontal spacing between subplots

plt.show()
```

### 5. How do train accidents contribute to spending? 

```python
# Selected company
selected_company = [
    'New Jersey Transit Rail Operations'
]

# Reapplying the filter to the original dataset for the selected company
selected_company_data = cleaned_df[cleaned_df['Reporting Railroad Name'].isin(selected_company)]

# Filtering for the top 10 causes for the selected company
top_causes_selected_company_reduced = selected_company_data['Total Damage Cost'].value_counts().head(10).index.tolist()
filtered_data_selected_company_reduced.loc[:, 'Total Damage Cost'] = pd.to_numeric(filtered_data_selected_company_reduced['Total Damage Cost'], errors='coerce')
```

```python
# Converting 'Total Damage Cost' to a numeric type, handling non-numeric values
filtered_data_selected_company_reduced['Total Damage Cost'] = pd.to_numeric(filtered_data_selected_company_reduced['Total Damage Cost'], errors='coerce')

# Redefining the bins for 'Total Damage Cost'
cost_bins = pd.cut(filtered_data_selected_company_reduced['Total Damage Cost'], 
                   bins=[0, 50000, 100000, 500000, 1000000, 4202000], 
                   labels=['0-50K', '50K-100K', '100K-500K', '500K-1M', '1M+'])

# Aggregating the data for the heatmap
heatmap_cost_data_company = filtered_data_selected_company_reduced.groupby([cost_bins, 'Primary Accident Cause']).size().unstack(fill_value=0)

# Creating the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_cost_data_company, annot=True, fmt="d", cmap='YlGnBu')
plt.title('Heatmap of Total Damage Cost vs. Primary Accident Cause (New Jersey Transit Rail Operations)')
plt.xlabel('Primary Accident Cause')
plt.ylabel('Total Damage Cost Range')
plt.xticks(rotation=45, ha='right')
plt.show()
```
